```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.pyplot import figure
```

```python
file = 'final_result.csv'
file = pd.read_csv(file)
file.values
```

```python
x = np.arange(10)
y = file.values

figure(figsize=(8, 6), dpi=80)

for index, row in enumerate(y):
    tick_spacing = 1
    plt.plot(x, row[2:12].astype(float), label=row[0])
    plt.xticks(np.arange(min(x), max(x)+1, 1.0))
plt.title('10-fold Cross Validation ROU-AUC score for all models')
plt.legend()
```

```python
name = []
values = []
color = ['blue', 'green', 'orange', 'red']
for index, row in enumerate(y):
    name.append(row[0])
    values.append(np.mean(row[2:12].astype(float)))
    
figure(figsize=(8, 6), dpi=80)
plt.title('Average ROC-AUC Score For All Models')
plt.bar(name, values, color=color)
```
